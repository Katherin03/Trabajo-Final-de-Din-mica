### Método: Segunda Ley de Newton

>Diagrama de cuerpo libre sobre el cilindro

>Por los grados de libertad con los que cuenta el sistema, realizamos la sumatoria de fuerzas alrededor del eje x y la sumatoria de torques en O alrededor del eje z.

>Sumatoria de fuerzas alrededor del eje x:
$$\Sigma F x=ma_x$$

$$mgsen(\phi)-fr=ma_{cm}$$

$$fr=mgsen(\phi)-ma_{cm}$$

$$\frac{1}{2}Ma_{cm}=mgsen(\phi)-ma_{cm}$$

$$\frac{1}{2}Ma_{cm}+Ma_{cm}=Mgsen(\phi)$$

$$\frac{3}{2}Ma_{cm}=Mgsen(\phi)$$

$$a_{cm}=\frac{2}{3}gsen(\phi)$$

>Sumatoria de torques alrededor del eje z, en el punto O:
$$\Sigma{\varGamma_O}=I_{cm}\alpha$$

$$Rfr=\frac{1}{2}MR^2\frac{a_{cm}}{R}$$

$$fr=\frac{1}{2}Ma_{cm}$$

>Considerando el sistema de coordenadas auxiliar, realizamos la sumatoria de torques alrededor del eje z, en el punto O':
$$\Sigma{\varGamma_{O'}}=I_{O'}\alpha$$

$$RMgsen(\theta)=I_{O'}\alpha$$

$$RMgsen(\theta)=\frac{1}{2}mR^2+mR^2\alpha$$

>Como $\alpha=\frac{a_{cm}}{R}$, tenemos:
$$\alpha=\frac{2}{3R}gsen(\phi)$$

>Por lo tanto, las ecuaciones del movimiento obtenidas a través del análisis a partir de la Segunda Ley de Newton, son:
$$a_{cm}=\frac{2}{3}gsen(\phi)$$

$$\alpha=\frac{2}{3R}gsen(\phi)$$

### Método: Lagrange

>Diagrama del cilindro

>Realizamos el análisis lagrangiano del sistema, considerando los grados de libertad del mismo y las energías involucradas.

$$ℒ=T-V$$

$$T=\frac{1}{2}M\dot{x}^2+\frac{1}{2}.\frac{1}{2}MR^2\dot{\theta}^2$$

$$V=-mgxsen(\phi)$$

$$ℒ=\frac{1}{2}M\dot{x}^2+\frac{1}{2}.\frac{1}{2}MR^2\dot{\theta}^2+mgxsen(\phi)$$

>Con respecto a $x$:
$$\frac{d}{dt}(\frac{\vartheta{ℒ}}{\vartheta\dot{x}})=\frac{d}{dt}(M\dot{x})=M\ddot{x}$$

$$\frac{\vartheta{ℒ}}{\vartheta{x}}=mgsen(\phi)$$

$$M\ddot{x}-mgsen(\phi)=\lambda \frac{\vartheta{F}}{\vartheta{x}}$$

>Con respecto a $\theta$:
$$\frac{d}{dt}(\frac{\vartheta{ℒ}}{\vartheta\dot{\theta}})=\frac{d}{dt}(\frac{MR^2\dot{\theta}}{2})=\frac{MR^2\ddot{\theta}}{2}$$

$$\frac{\vartheta{ℒ}}{\vartheta{\theta}}=0$$

$$\frac{MR^2\ddot{\theta}}{2}=\lambda \frac{\vartheta{F}}{\vartheta{\theta}}$$

$$F=x-R\theta=0$$

$$\frac{\vartheta{F}}{\vartheta{x}}=1$$

$$\frac{\vartheta{F}}{\vartheta{\theta}}=-R$$

>Reemplazamos en las ecuaciones encontradas anteriormente:

$$M\ddot{x}-Mgsen(\phi)=\lambda$$

$$\frac{MR^2\ddot{\theta}}{2}=-\lambda R$$

$$\frac{MR\ddot{\theta}}{2}=-\lambda$$

$$\frac{1}{2}MR\ddot{\theta}=\lambda$$

>Ya que tenemos más incógnitas que ecuaciones, procedemos a derivar la ecuación de la fuerza encontrada previamente:

$$x=R\theta$$

$$\dot{x}=R\dot{\theta}$$

$$\ddot{x}=R\ddot{\theta}$$

>Reemplazamos las ecuaciones encontradas:
$$-\frac{1}{2}M\ddot{x}=\lambda$$

$$-\frac{1}{2}M\ddot{x}=M\ddot{x}-Mgsen(\phi)$$

$$\ddot{x}+\frac{1}{2}\ddot{x}=gsen(\phi)$$

$$\frac{3}{2}\ddot{x}=gsen(\phi)$$

$$\ddot{x}=\frac{2}{3}gsen(\phi)$$

>Finalmente reemplazamos la última ecuación:
$$-\frac{1}{2}M\frac{2}{3}gsen(\phi)=\lambda$$

$$\lambda=-\frac{M}{3}gsen(\phi)$$

$$-\frac{M}{3}gsen(\phi)=-\frac{1}{2}MR\ddot{\theta}$$

$$\ddot{\theta}=\frac{2}{3R}gsen(\phi)$$

>Por lo tanto, las ecuaciones del movimiento obtenidas a través del análisis a partir del método de Lagrange, son:
$$\ddot{x}=\frac{2}{3}gsen(\phi)$$

$$\ddot{\theta}=\frac{2}{3R}gsen(\phi)$$
